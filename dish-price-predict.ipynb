```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score, max_error
from sklearn.preprocessing import StandardScaler
from colorama import Fore, Style
from sklearn.utils import resample
import spacy

nlp = spacy.load("en_core_web_sm")

def preprocess_text_advanced(text):
    doc = nlp(text)
    return " ".join([token.lemma_ for token in doc if not token.is_stop and token.is_alpha])

def engineer_features(menu_data, vectorizer=None, scaler=None):
    if vectorizer is None:
        vectorizer = TfidfVectorizer(stop_words='english', preprocessor=preprocess_text_advanced)
    X_text = vectorizer.fit_transform([item['name'] + ' ' + item['description'] for item in menu_data])

    X_additional = np.array([[len(item['description']), len(item['name'])] for item in menu_data])
    if scaler is None:
        scaler = StandardScaler()
    X_additional_scaled = scaler.fit_transform(X_additional)
    X_combined = np.hstack((X_text.toarray(), X_additional_scaled))

    return X_combined, vectorizer, scaler

def train_model(X, y, model_type='ridge', alpha=0.1):
    models = {
        'ridge': Ridge(alpha=alpha),
        'lasso': Lasso(alpha=alpha),
        'random_forest': RandomForestRegressor(n_estimators=100, random_state=42),
        'gradient_boosting': GradientBoostingRegressor(n_estimators=100, random_state=42),
        'knn': KNeighborsRegressor(),
        'svm': SVR()
    }

    model = models.get(model_type)
    if model is None:
        raise ValueError(f"Invalid model_type: {model_type}")

    model.fit(X, y)
    return model

def train_model_bootstrap(X, y, model_type='ridge', alpha=0.1, param_grid=None, num_iterations=1):
    models_params = {
        'ridge': {'model': Ridge(alpha=alpha), 'param_grid': None},
        'lasso': {'model': GridSearchCV(Lasso(), {'alpha': [0.001]}, scoring='neg_mean_squared_error', cv=5), 'param_grid': None},
        'random_forest': {'model': RandomForestRegressor(n_estimators=100, random_state=42), 'param_grid': None},
        'gradient_boosting': {'model': GradientBoostingRegressor(n_estimators=100, random_state=42), 'param_grid': None},
        'knn': {'model': GridSearchCV(KNeighborsRegressor(), {'n_neighbors': [3, 5, 7, 9], 'weights': ['uniform', 'distance']}, scoring='neg_mean_squared_error', cv=5), 'param_grid': None},
        'svm': {'model': GridSearchCV(SVR(), {'C': [0.1, 1, 10], 'gamma': [0.001, 0.01, 0.1, 1]}, scoring='neg_mean_squared_error', cv=5), 'param_grid': None}
    }

    model_info = models_params.get(model_type)
    if model_info is None:
        raise ValueError(f"Invalid model_type: {model_type}")

    model = model_info['model']
    if param_grid is not None:
        model_info['param_grid'] = param_grid
        model = GridSearchCV(model, param_grid, scoring='neg_mean_squared_error', cv=5)

    mse_bootstrap = []

    for _ in range(num_iterations):
        X_boot, y_boot = resample(X, y, random_state=42)
        model.fit(X_boot, y_boot)
        mse, _, _, _ = evaluate_model(model, X, y)
        mse_bootstrap.append(mse)

    avg_mse_bootstrap = np.mean(mse_bootstrap)

    return avg_mse_bootstrap

def evaluate_model(model, X, y):
    """Evaluate regression model and return metrics."""
    predictions = model.predict(X)

    mse = mean_squared_error(y, predictions)
    mae = mean_absolute_error(y, predictions)
    r2 = r2_score(y, predictions)
    evs = explained_variance_score(y, predictions)
    max_err = max_error(y, predictions)

    return mse, mae, r2, max_err

def print_model_metrics(model_type, mse, mae, r2, max_err):
    print(f"{Fore.BLUE}Model: {model_type}{Style.RESET_ALL}")
    print(f"{Fore.GREEN}Mean Squared Error: {mse:.2f}{Style.RESET_ALL}")
    print(f"{Fore.GREEN}Mean Absolute Error: {mae:.2f}{Style.RESET_ALL}")
    print(f"{Fore.GREEN}R-squared: {r2:.2f}{Style.RESET_ALL}")
    print(f"{Fore.GREEN}Max Error: {max_err:.2f}{Style.RESET_ALL}")
    print("=" * 75)

def main(num_iterations=1):

    menu_df = pd.DataFrame(menu_data)

    y = menu_df['price'].values

    X, vectorizer, scaler = engineer_features(menu_data)
    X_all, _, y_all, _ = train_test_split(X, y, test_size=0.2, shuffle=True)

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)

    models_to_try = ['ridge', 'lasso', 'random_forest', 'gradient_boosting', 'knn', 'svm']

    predicted_prices = np.zeros(len(models_to_try))

    for model_type_value in models_to_try:
        alpha_value = 0.001 if model_type_value == 'lasso' else 0.1

        avg_mse_bootstrap = train_model_bootstrap(X_train, y_train, model_type=model_type_value, alpha=alpha_value, num_iterations=num_iterations)

        print(f"{model_type_value} - Average MSE across {num_iterations} bootstrap samples: {Fore.YELLOW}{avg_mse_bootstrap:.2f}{Style.RESET_ALL}")

        final_model = train_model(X_train, y_train, model_type=model_type_value, alpha=alpha_value)

        mse, mae, r2, max_err = evaluate_model(final_model, X_test, y_test)
        print_model_metrics(model_type_value, mse, mae, r2, max_err)

        new_dish = {"name": "Margherita Pizza", "description": "Traditional Italian pizza with tomato and mozzarella", "price": None}
        new_dish_text = [new_dish['name'] + ' ' + new_dish['description']]
        new_dish_features = np.hstack((vectorizer.transform(new_dish_text).toarray(), scaler.transform([[len(new_dish['description']), len(new_dish['name'])]])))
        predicted_price = final_model.predict(new_dish_features)
        predicted_prices[models_to_try.index(model_type_value)] = predicted_price[0]

        print(f"{model_type_value} - Predicted Price for '{new_dish['name']}': ${predicted_price[0]:.2f}")
        print("=" * 75)

    average_predicted_price = np.mean(predicted_prices)

    print("Predicted Prices from all the models side by side:")
    for model_type, price in zip(models_to_try, predicted_prices):
        print(f"{model_type}: ${price:.2f}")

    print("=" * 75)
    print(f"Average Predicted Price for the new dish '{new_dish['name']}': ${Fore.RED}{average_predicted_price:.2f}{Style.RESET_ALL}")

if __name__ == "__main__":
    main(num_iterations=50)
```
